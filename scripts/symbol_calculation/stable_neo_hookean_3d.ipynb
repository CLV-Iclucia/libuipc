```python
import sys
import os
sys.path.append('../')
import pathlib as pl
from SymEigen import *
from sympy import symbols
from project_dir import project_dir
Gen = EigenFunctionGenerator()
Gen.MacroBeforeFunction("__host__ __device__")

def I2(VecF):
    return VecF.dot(VecF)
    
def I3(F):
    return F.det()
```

```python
mu, lmd = symbols("mu lambda")
F = Eigen.Matrix("F", 3, 3)
F
```

```python
VecF = F.Vectorize("VecF")
VecF
```

```python
I2 = I2(VecF)
I2
```

```python
I3 = I3(F)
I3
```

```python
SNK = (mu/2*(I2-3) - mu * (I3-1) + lmd/2*(I3-1)**2) + (mu/lmd)**2
SNK
```

```python
dSNKdVecF = VecDiff(SNK, VecF)
dSNKdVecF
```

```python
ddSNKddVecF = VecDiff(dSNKdVecF, VecF)
ddSNKddVecF
```

```python
Cl = Gen.Closure(mu, lmd, VecF)
```

```python
s = f'''
{Cl("SNK",SNK)}
{Cl("dSNKdVecF",dSNKdVecF)}
{Cl("ddSNKddVecF",ddSNKddVecF)}
'''
print(s)

f = open( project_dir() / 'src/uipc/backends\cuda/finite_element/constitutions/sym/stable_neo_hookean_3d.inl', 'w')
f.write(s)
f.close()
```
